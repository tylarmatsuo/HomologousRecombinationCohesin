```python
# Plot of ratios for first passage rel. change

# For plotting proportion achieving passage within a certain timeframe
EquilibriationOffset = 1030

TimestepCutoff = 5400 + EquilibriationOffset

a_data = np.load("/path/to/first_pass_time_left.npy") 
b_data = np.load("/path/to/first_pass_time_left.npy")

a_valid_idx = np.where(np.all(a_data, axis=0))[0]
b_valid_idx = np.where(np.all(b_data, axis=0))[0]

a_means = np.mean(a_data <= TimestepCutoff, axis=0)
b_means = np.mean(b_data <= TimestepCutoff, axis=0)

Na = a_data.shape[0]
Nb = b_data.shape[0]

a_sem = np.sqrt(a_means * (1 - a_means) / Na)
b_sem = np.sqrt(b_means * (1 - b_means) / Nb)

rel_change_a_b_sem = 100 * np.sqrt(
    (a_sem / b_means)**2 +
    (a_means * b_sem / b_means**2)**2
)

rel_change_a_b = (-100) * (1 - a_means/b_means)

distances_of_interest = [0, 440, 560]
no_offset_homology_pos = 3000

rel_means = []
rel_sems  = []

for d in distances_of_interest:
    vals = []
    errs = []

    if d == 0:
        pos = no_offset_homology_pos
        vals.append(rel_change_a_b[pos])
        errs.append(rel_change_a_b_sem[pos])
    else:
        for pos in (no_offset_homology_pos - d,
                    no_offset_homology_pos + d):
            if 0 <= pos < len(rel_change_a_b):
                vals.append(rel_change_a_b[pos])
                errs.append(rel_change_a_b_sem[pos])

    rel_means.append(np.nanmean(vals))
    rel_sems.append(
        np.sqrt(np.nansum(np.array(errs)**2)) / len(errs)
    )

plt.figure()
plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'axes.titlesize': 26})
plt.rcParams.update({'legend.fontsize': 20})

plt.bar(
    [str(d) for d in distances_of_interest],
    rel_means,
    yerr=rel_sems,
    capsize=4,
    color = 'xkcd:greeny blue'
)
plt.axhline(0)
plt.xlabel("Distance from DSB (kb)")
plt.ylabel("∆ HR Efficacy")
plt.ylim(-120, 0)
plt.title("∆LEF (60 Min, w/ Clamp)")
plt.gca().get_yaxis().set_visible(True) 
plt.tight_layout()
plt.show()
```

```python
# Weibull 2 parameter fit
a_data = np.load("/path/to/first_pass_time_left.npy")
b_data = np.load("/path/to/first_pass_time_left.npy")
c_data = np.load("/path/to/first_pass_time_left.npy")
d_data = np.load("/path/to/first_pass_time_left.npy")

# Weibull CDF: A * (1 - exp(-(x/lam)^k))
# plateau is fixed at 1
def weibull_cdf(x, k, lam):
    return 1 - np.exp(-(x / lam)**k)

EquilibriationOffset = 1030

x_values_a, y_values_a = [], []
x_values_b, y_values_b = [], []
x_values_c, y_values_c = [], []
x_values_d, y_values_d = [], []

for i in range(20000):
    x = i * 2/3
    x_values_a.append(x); x_values_b.append(x)
    x_values_c.append(x); x_values_d.append(x)
    y_values_a.append(np.sum(a_data[:,3000] < i + EquilibriationOffset) / len(a_data))
    y_values_b.append(np.sum(b_data[:,3000] < i + EquilibriationOffset) / len(b_data))
    y_values_c.append(np.sum(c_data[:,3000] < i + EquilibriationOffset) / len(c_data))
    y_values_d.append(np.sum(d_data[:,3000] < i + EquilibriationOffset) / len(d_data))

datasets = [
    (x_values_a, y_values_a, 'xkcd:ruby',          'No LEFs'),
    (x_values_b, y_values_b, 'xkcd:ochre',          'Yes LEFs, No Clamp'),
    (x_values_c, y_values_c, 'mediumseagreen',       'Yes LEFs, No Anchor'),
    (x_values_d, y_values_d, 'xkcd:dark navy blue', 'Yes LEFs, Yes Clamp & Anchor'),
]

x_fit = np.linspace(0, max(x_values_a), 1000)

print("=== Fit Parameters (Weibull CDF) ===")
plt.figure(figsize=(7.0, 5.5))

for x_vals, y_vals, color, label in datasets:
    x_arr = np.array(x_vals)
    y_arr = np.array(y_vals)

    p0     = [0.5,  1000.0]

    popt, pcov = curve_fit(weibull_cdf, x_arr, y_arr, p0=p0, maxfev=20000)
    k, lam = popt
    perr = np.sqrt(np.diag(pcov))

    # Median: solve A*(1 - exp(-(t/lam)^k)) = 0.5*A  →  t = lam * ln(2)^(1/k)
    t_50 = lam * np.log(2)**(1/k)

    print(f"\n{label}")
    print(f"  A (plateau)          = Fixed at 1.0")
    print(f"  k (shape)            = {k:.4f} ± {perr[0]:.4f}")
    print(f"  lam (char. time)     = {lam:.2f} ± {perr[1]:.2f} s")
    print(f"  Median FPT (t_50)    = {t_50:.2f} s")

    plt.plot(x_fit, weibull_cdf(x_fit, *popt), color=color, linewidth=2, linestyle='--')
    plt.plot(x_vals, y_vals, color=color, alpha=0.6, linewidth=3, label=label)

plt.rcParams.update({'font.size': 22})
plt.rcParams.update({'axes.titlesize': 24})
plt.title('Cohesin and Search Kinetics')
plt.xlabel('Time (s)')
plt.ylabel('Prop. Homology Found')
plt.xscale('linear')
plt.yscale('linear')
#plt.legend(reverse=True)
plt.tight_layout()
plt.show()
```

```python
# Weibull 2 parameter fit for ends disconnected case

from scipy.special import gamma as gamma_fn

def weibull_cdf(x, k, lam):
    return 1 - np.exp(-(x / lam)**k)

# For untethered: both left and right must be contacted by timestep i
def frac_both_contacted(left_data, right_data, i, equilibration_offset):
    cutoff = i + equilibration_offset
    both = (left_data <= cutoff) & (right_data <= cutoff)
    return np.sum(both) / len(left_data)

EquilibriationOffset = 1030

a_left_data  = np.load("/path/to/first_pass_time_left.npy")
a_right_data = np.load("/path/to/first_pass_time_right.npy")

b_left_data  = np.load("/path/to/first_pass_time_left.npy")
b_right_data = np.load("/path/to/first_pass_time_right.npy")

c_left_data  = np.load("/path/to/first_pass_time_left.npy")
c_right_data = np.load("/path/to/first_pass_time_right.npy")

d_left_data  = np.load("/path/to/first_pass_time_left.npy")
d_right_data = np.load("/path/to/first_pass_time_right.npy")

x_values_a, y_values_a = [], []
x_values_b, y_values_b = [], []
x_values_c, y_values_c = [], []
x_values_d, y_values_d = [], []

for i in range(20000):
    x = i * 2/3
    x_values_a.append(x); x_values_b.append(x)
    x_values_c.append(x); x_values_d.append(x)
    y_values_a.append(frac_both_contacted(a_left_data[:,3000], a_right_data[:,3000], i, EquilibriationOffset))
    y_values_b.append(frac_both_contacted(b_left_data[:,3000], b_right_data[:,3000], i, EquilibriationOffset))
    y_values_c.append(frac_both_contacted(c_left_data[:,3000], c_right_data[:,3000], i, EquilibriationOffset))
    y_values_d.append(frac_both_contacted(d_left_data[:,3000], d_right_data[:,3000], i, EquilibriationOffset))

datasets = [
    (x_values_a, y_values_a, 'xkcd:ruby',          'No LEFs'),
    (x_values_b, y_values_b, 'xkcd:ochre',      'Yes LEFs, No Clamp'),
    (x_values_c, y_values_c, 'mediumseagreen',          'Yes LEFs, No Anchor'),
    (x_values_d, y_values_d, 'xkcd:dark navy blue', 'Yes LEFs, Yes Clamp & Anchor'),
]

x_fit = np.linspace(0, max(x_values_a), 1000)

plt.rcParams.update({'font.size': 22, 'legend.fontsize': 19, 'axes.titlesize': 24})
print("=== Fit Parameters (Weibull CDF, Untethered) ===")
plt.figure(figsize=(10, 7))

for x_vals, y_vals, color, label in datasets:
    x_arr = np.array(x_vals)
    y_arr = np.array(y_vals)

    popt, pcov = curve_fit(weibull_cdf, x_arr, y_arr, p0=[0.5, 1000.0], maxfev=20000)
    k, lam = popt
    perr = np.sqrt(np.diag(pcov))

    t_50     = lam * np.log(2)**(1/k)
    mean_fpt = lam * gamma_fn(1 + 1/k)

    print(f"\n{label}")
    print(f"  A (plateau)       = Fixed at 1.0")
    print(f"  k (shape)         = {k:.4f} ± {perr[0]:.4f}")
    print(f"  lam (char. time)  = {lam:.2f} ± {perr[1]:.2f} s")
    print(f"  Median FPT (t_50) = {t_50:.2f} s")
    print(f"  Mean FPT          = {mean_fpt:.2f} s")

    plt.plot(x_fit, weibull_cdf(x_fit, *popt), color=color, linewidth=2, linestyle='--')
    plt.plot(x_vals, y_vals, color=color, alpha=0.6, linewidth=3, label=label)

plt.title('Cohesin Accelerates Untethered Searches')
plt.xlabel('Time (s)')
plt.ylabel('Prop. Homology Found')
plt.xscale('linear')
plt.yscale('linear')
plt.legend(reverse=True)
plt.tight_layout()
plt.show()
```

```python
# Plotting correlation between TAD size and passage times

# exponential CDF, plateau at 1
def exp_cdf(x, lam):
    return 1 - np.exp(-x / lam)

def fit_t50(data, position, offset, timesteps=10000):
    """
    Fit a 1-parameter exponential CDF to FPT data at a single position.
    Returns t_50 in seconds.
    """
    col   = data[:, position]
    x_arr = np.array([i * (2/3) for i in range(timesteps)])
    y_arr = np.array([np.sum(col < i + offset) / len(col) for i in range(timesteps)])

    popt, _ = curve_fit(exp_cdf, x_arr, y_arr, p0=[1000.0], maxfev=20000)
    lam, = popt
    t_50 = lam * np.log(2)
    return t_50

lef_data_paths = {
    "1MbTAD":   "/path/to/first_pass_time_left.npy",
    "900kbTAD": "/path/to/first_pass_time_left.npy",
    "800kbTAD": "/path/to/first_pass_time_left.npy",
    "700kbTAD": "/path/to/first_pass_time_left.npy",
    "600kbTAD": "/path/to/first_pass_time_left.npy",
    "500kbTAD": "/path/to/first_pass_time_left.npy",
    "400kbTAD": "/path/to/first_pass_time_left.npy",
    "300kbTAD": "/path/to/first_pass_time_left.npy",
    "200kbTAD": "/path/to/first_pass_time_left.npy",
    "100kbTAD": "/path/to/first_pass_time_left.npy",
}
lef_comparisons = [  # (data_key, TAD_size_kb)
    ("1MbTAD",   1000),
    ("900kbTAD",  900),
    ("800kbTAD",  800),
    ("700kbTAD",  700),
    ("600kbTAD",  600),
    ("500kbTAD",  500),
    ("400kbTAD",  400),
    ("300kbTAD",  300),
    ("200kbTAD",  200),
    ("100kbTAD",  100),
]

no_lef_data_paths = {
    "1MbTAD":   "/path/to/first_pass_time_left.npy",
    "800kbTAD": "/path/to/first_pass_time_left.npy",
    "600kbTAD": "/path/to/first_pass_time_left.npy",
    "500kbTAD": "/path/to/first_pass_time_left.npy",
    "400kbTAD": "/path/to/first_pass_time_left.npy",
    "200kbTAD": "/path/to/first_pass_time_left.npy",
}
no_lef_comparisons = [  # (data_key, TAD_size_kb)
    ("1MbTAD",   1000),
    ("800kbTAD",  800),
    ("600kbTAD",  600),
    ("500kbTAD",  500),
    ("400kbTAD",  400),
    ("200kbTAD",  200),
]

# Extra cases
# Each dict: label, color, data_paths {key: path}, comparisons [(key, TAD_size_kb)]
extra_cases = [
    {
            "label": "Loops, No DSB-Interaction",
            "color": "mediumseagreen",
            "data_paths": {
                "1MbTAD":   "/path/to/first_pass_time_left.npy",
                "800kbTAD": "/path/to/first_pass_time_left.npy",
                "600kbTAD": "/path/to/first_pass_time_left.npy",
                "500kbTAD": "/path/to/first_pass_time_left.npy",
                "400kbTAD": "/path/to/first_pass_time_left.npy",
                "200kbTAD": "/path/to/first_pass_time_left.npy",
            },
            "comparisons": [
                ("1MbTAD",   1000),
                ("800kbTAD",  800),
                ("600kbTAD",  600),
                ("500kbTAD",  500),
                ("400kbTAD",  400),
                ("200kbTAD",  200),
            ],
        }
]

offset        = 1030
fit_position  = 2100   # genomic position column to fit
timesteps     = 10000  # number of CDF evaluation points

# Shared TAD sizes for regression
shared_sizes  = {200, 400, 500, 600, 800, 1000}

cases = [
    {"label": "Loops, Yes DSB-Interaction",    "color": "xkcd:dark navy blue", "data_paths": lef_data_paths,    "comparisons": lef_comparisons},
    *extra_cases,
    {"label": "No Loops", "color": "xkcd:ruby",   "data_paths": no_lef_data_paths, "comparisons": no_lef_comparisons},
]

print(f"Fitting Exponential CDFs at position {fit_position}...\n")
for case in cases:
    case["t50_by_size"] = {}  # TAD_size_kb → t_50 (s)
    for key, TAD_size in case["comparisons"]:
        data = np.load(case["data_paths"][key])
        t50  = fit_t50(data, fit_position, offset, timesteps)
        case["t50_by_size"][TAD_size] = t50
        print(f"  [{case['label']}]  {TAD_size:>5} kb  →  t_50 = {t50:.2f} s")
    print()

def regression_on_shared(case, shared_sizes):
    pairs = [(sz, t50) for sz, t50 in case["t50_by_size"].items() if sz in shared_sizes]
    pairs.sort()
    x = np.array([p[0] for p in pairs])
    y = np.array([p[1] for p in pairs])
    slope, intercept, r_value, p_value, _ = stats.linregress(x, y)
    return x, y, slope, intercept, r_value, p_value

fig, ax = plt.subplots(figsize=(7, 5))

plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'legend.fontsize': 20})
plt.rcParams.update({'axes.titlesize': 24})

all_stats = []
eq_lines  = []

for case in cases:
    x, y, slope, intercept, r_value, p_value = regression_on_shared(case, shared_sizes)

    shared_mask = np.isin(x, list(shared_sizes))
    ax.scatter(x[shared_mask],  y[shared_mask],  color=case["color"], s=55,
               alpha=0.8, zorder=3, label=case["label"])
    ax.scatter(x[~shared_mask], y[~shared_mask], color=case["color"], s=55,
               alpha=0.4, zorder=3, marker='o', facecolors='none',
               edgecolors=case["color"], linewidths=1.5)

    ax.plot(x, slope * x + intercept, color=case["color"],
            linewidth=2, linestyle='--', alpha=0.85)

    all_stats.append((case["label"], slope, intercept, r_value, p_value))
    sign = "+" if intercept >= 0 else "-"
    eq_lines.append(
        f"{case['label']:>8}:  y = {slope:.2f}x {sign} {abs(intercept):.1f}"
        f"   R = {r_value:.4f}  p = {p_value:.4e}"
    )

if False:
    ax.text(
        0.05, 0.95, "\n".join(eq_lines),
        transform=ax.transAxes, fontsize=9, verticalalignment='top',
        fontfamily='monospace',
        bbox=dict(boxstyle='round', facecolor='xkcd:light gray', alpha=0.8),
    )

ax.set_title("Search Times by TAD Size")
ax.set_xlabel("TAD Size (kb)")
ax.set_ylabel(f"$t_{{50}}$ (s)")
ax.legend(reverse=True, loc='upper left', borderpad=0.2, fontsize=14)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

col_w  = 14
header = f"{'':12}" + "".join(f"{s[0]:>{col_w}}" for s in all_stats)
print("=" * len(header))
print(header)
print("=" * len(header))
for row_label, fmt, getter in [
    ("Slope",     ".4f", lambda s: s[1]),
    ("Intercept", ".2f", lambda s: s[2]),
    ("R",         ".4f", lambda s: s[3]),
    ("R²",        ".4f", lambda s: s[3]**2),
    ("p-value",   ".4e", lambda s: s[4]),
]:
    print(f"{row_label:<12}" + "".join(f"{getter(s):{col_w}{fmt}}" for s in all_stats))
print("=" * len(header))
```
